```python
# read and open the dataset
data = pd.read_csv('PriceCheckInput.csv')
data
```

```python
# requirements for validations 
# Let used the for loop to compare variables.
for i in data:

    d=(((data.Low <= data.Close) & (data.Close <= data.High)) 
       & ((data.Low <= data.Open) & (data.Open <=data.High)))
    

data["IsValid"]= d    
```

# Moving Average

```python
# Let use the pandas in-built rolling function which allow us to calculate the moving average
data['MovingAvg'] = data["Close"].rolling(window=5).mean()
```
